```python
import json
import pandas as pd
import argparse
```

```python
with open('level1_coco.json', 'r') as f1:
    anno1 = json.load(f1)
with open('level1_2_coco.json', 'r') as f2:
    anno2 = json.load(f2)
faultyList = []
```

```python
# Create a data frame
df1 = pd.DataFrame(columns = ['imgID', 'imgPath', 'dimensions', 'areas', 'labels', 'polygons', 'bboxes'])

# fill the df1 with entire anno data
for i in range(len(anno1['images'])):
  tempDict = {'imgID': anno1['images'][i]['id'],
            'imgPath':anno1['images'][i]['file_name'],
            'dimensions': (anno1['images'][i]['height'],anno1['images'][i]['width']),
            'areas': [],
            'labels':[],
            'polygons':[],
            'bboxes':[]}
  df1 = df1.append(tempDict, ignore_index=True)

df1.set_index('imgID', inplace=True)

for i in range(len(anno1['annotations'])): 
  locIndex = int(anno1['annotations'][i]['image_id'])
  try:
    df1.loc[locIndex]['areas'].append(anno1['annotations'][i]['area'])
    df1.loc[locIndex]['labels'].append(anno1['annotations'][i]['category_id'])
    df1.loc[locIndex]['polygons'].append(anno1['annotations'][i]['segmentation'])
    df1.loc[locIndex]['bboxes'].append(anno1['annotations'][i]['bbox'])
  except:
    faultyList.append(locIndex)

df1.sort_index(inplace=True)
```

```python
# Create a data frame
df2 = pd.DataFrame(columns = ['imgID', 'imgPath', 'dimensions', 'areas', 'labels', 'polygons', 'bboxes'])

# fill the df2 with entire anno data
for i in range(len(anno2['images'])):
  tempDict = {'imgID': anno2['images'][i]['id'],
            'imgPath':anno2['images'][i]['file_name'],
            'dimensions': (anno2['images'][i]['height'],anno2['images'][i]['width']),
            'areas': [],
            'labels':[],
            'polygons':[],
            'bboxes':[]}
  df2 = df2.append(tempDict, ignore_index=True)

df2.set_index('imgID', inplace=True)

for i in range(len(anno2['annotations'])): 
  locIndex = int(anno2['annotations'][i]['image_id'])
  try:
    df2.loc[locIndex]['areas'].append(anno2['annotations'][i]['area'])
    df2.loc[locIndex]['labels'].append(anno2['annotations'][i]['category_id'])
    df2.loc[locIndex]['polygons'].append(anno2['annotations'][i]['segmentation'])
    df2.loc[locIndex]['bboxes'].append(anno2['annotations'][i]['bbox'])
  except:
    faultyList.append(locIndex)

df2.sort_index(inplace=True)
```

```python
df2
```

```python
df = pd.concat([df1, df2], ignore_index=True)
for id, row in df.iterrows():
    if len(row['bboxes']) == 0:
        df.drop(id, axis=0, inplace=True)
```

```python
combinedJson = {
    'annotations': [],
    'categories': anno1['categories'],
    'images': [],
    'info': {},
    'licenses': []
}
annoIDCount = 0
for imgID, row in df.iterrows():
    tempImages = {
      "file_name": row['imgPath'],
      "height": row['dimensions'][0],
      "width": row['dimensions'][1],
      "id": imgID,
    }
    combinedJson['images'].append(tempImages)
    for i in range(len(row['labels'])):
      tempAnno = {
        "area": row['areas'][i],
        "bbox": row['bboxes'][i],
        "category_id": row['labels'][i],
        "id": annoIDCount,
        "image_id": imgID,
        "iscrowd": 0,
        "segmentation": [row['polygons'][i]]
      }
      combinedJson['annotations'].append(tempAnno)
      annoIDCount += 1

with open("CombinedCoco.json", "w") as outfile:
    json.dump(combinedJson, outfile)
```
